# Capabilities
The core Capabilities of the Modeling Environment derived from the Requirements:

```python
from utilities import *
import pandas as pd

pd.set_option('display.max_colwidth', None)

df = dataframe("capabilities.json")
cap_req = (
    df
    .groupby(["cap_id", "cap_desc"])["req_id"]
    .apply(
        lambda ids: ", ".join(
            sorted(
                {str(x) for x in ids if pd.notna(x)}
            )
        )
    )
    .reset_index(name="requirements")
)

cap_req
```
